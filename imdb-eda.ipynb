```python
# basic libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go
import warnings
import nltk
from nltk.corpus import stopwords
from wordcloud import WordCloud
sns.set(style="darkgrid")
warnings.filterwarnings('ignore')
```

<h2>Cleaning the Data<h2>

```python
# loading dataset
data=pd.read_csv('/kaggle/input/imdb-top-100-movies/movies.csv')
df=data.copy()
df.head()
```

```python
# droping index column
df.drop('index',axis=1,inplace=True)
```

```python
# check missing values
df.isnull().sum().sort_values(ascending = False)
```

```python
df[df['gross_total'].isnull()]
```

```python
# fill missing values with 0
df.fillna(0,inplace=True)
df.isnull().sum()
```

<h1>Data Visualization<h1>

<h2>Visualization of Rating Distribution<h2>

```python
plt.figure(figsize = (10, 8))
sns.countplot(data=df, x="imdb_rating")
plt.title('Movies vs Rating', fontdict={'size':20,'color':'darkblue'})
plt.xlabel('Imdb Rating', fontdict={'size':20,'color':'darkblue'})
plt.xticks(rotation=90)
plt.ylabel('Count',fontdict={'size':20,'color':'darkblue'})
sns.color_palette("pastel")
plt.show()
```

<h2>Yearwise Distribution<h2>

```python
# cleaning the year of release and changing the data type to int
df['year_of_release'] = df['year_of_release'].str.replace('(', '')
df['year_of_release'] = df['year_of_release'].str.replace(')', '')

df['year_of_release'] = df['year_of_release'].astype('int64')
```

```python
plt.figure(figsize=(20,5))
sns.set(style="darkgrid")
plt.title('Yearwise Distribution',fontdict={'size':20,'color':'darkblue'})
ax = sns.countplot(x="year_of_release", data=df)
plt.xlabel('Year', fontdict={'size':20,'color':'darkblue'})
plt.xticks(rotation=90)
plt.ylabel('No of Movies',fontdict={'size':20,'color':'darkblue'})
plt.show()
```

<h2>Vizualization of Most Common genres<h2>

```python
def top_genre_pie_chart(top):
    fig = plt.figure(figsize =(10, 9))
    plt.pie(df['genre'].value_counts().head(top), labels = df['genre'].head(top), autopct='%1.2f%%')
    plt.title(f'Top {top} most common genres',fontdict={'size':20,'color':'darkblue'})
    plt.show()

#top value limit 55
top_genre_pie_chart(10)
```

<h2>Category Distribution<h2>

```python
plt.figure(figsize=(10,8))
sns.set(style="darkgrid")
sns.countplot(x="category", data=df, palette="Set2", order=df['category'].value_counts().index[0:15])
plt.title('Category Distribution',fontdict={'size':20,'color':'darkblue'})
plt.xlabel('Category', fontdict={'size':20,'color':'darkblue'})
plt.xticks(rotation=90)
plt.ylabel('No of Categories',fontdict={'size':20,'color':'darkblue'}) 
plt.show()
```

<h2>Category vs Gross total<h2>

```python
# cleaning the gross tota and changing the data type to float
df['gross_total'] = df['gross_total'].str.replace('$', '')
df['gross_total'] = df['gross_total'].str.replace('M', '')

df['gross_total']=df['gross_total'].astype('float64')
```

```python
fig, ax= plt.subplots(figsize=(20,6))
ax = sns.stripplot(y="category", x="gross_total",data=df, palette='pastel',
                   size = 10, marker='D', linewidth=1)
plt.xticks(rotation=90)
plt.title('Category vs Gross Total',fontdict={'size':20,'color':'darkblue'})
plt.xlabel("Gross Total('M)",fontdict={'size':20,'color':'darkblue'})
plt.ylabel('Category',fontdict={'size':20,'color':'darkblue'})
plt.show()
```

<h2>Category vs Rating<h2>

```python
fig, ax=plt.subplots(figsize=(15,10))
plt.xticks(rotation=60)
ax = sns.boxplot(x="category", y="imdb_rating", data=df)
plt.title('Category vs Rating',fontdict={'size':20,'color':'darkblue'})
plt.xlabel("Category",fontdict={'size':20,'color':'darkblue'})
plt.ylabel('Imdb Rating',fontdict={'size':20,'color':'darkblue'})
plt.show()
```

<h2>Run Time Distribution<h2>

```python
# cleaning the run time and changing the data type to int
df['run_time'] = df['run_time'].str.replace('min', '')
df['run_time'] = df['run_time'].astype('int64')
```

```python
sns.displot(df, x= 'run_time', aspect=1.9)
plt.title('Run time Distribution',fontdict={'size':20,'color':'darkblue'})
plt.xlabel("Run time('min')",fontdict={'size':20,'color':'darkblue'})
plt.ylabel("Count",fontdict={'size':20,'color':'darkblue'})
plt.show()
```
